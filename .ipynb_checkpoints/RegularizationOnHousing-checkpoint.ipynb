```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df
```

```python
df = pd.get_dummies(df,drop_first = True).astype(int)
df
```

```python
X = df.drop(["price"],axis = 1)
X.head()
```

```python
y = df.price
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.3,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(Xtrain,ytrain)
```

```python
reg.score(Xtest,ytest)
```

```python
reg.score(Xtrain,ytrain)
```

# As we can see training data set has more accuracy than testing data set there is overfitting
# L1 Regularization Lasso

```python
from sklearn import linear_model

lassoReg = linear_model.Lasso(alpha=50,max_iter = 100,tol = 0.1)
lassoReg.fit(Xtrain,ytrain)
```

```python
lassoReg.score(Xtest,ytest)
```

```python
lassoReg.score(Xtrain,ytrain)
```

# L2 Regularization

```python
from sklearn.linear_model import Ridge
l2 = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
l2.fit(Xtrain,ytrain)
```

```python
l2.score(Xtest,ytest)
```

```python
l2.score(Xtest,ytest)
```

#### Parameters:
#### alpha = 50
#### Controls regularization strength.
#### Higher value = stronger penalty = smaller coefficients = less overfitting.
#### max_iter = 100
#### Maximum number of iterations the algorithm can run to find best weights.
#### tol = 0.1
#### Tolerance for stopping.
#### If improvement becomes smaller than 0.1, it stops early.
#### Simple Meaning:

#### “Create a Ridge model with strong regularization, allow 100 steps, stop when changes are very small.”
